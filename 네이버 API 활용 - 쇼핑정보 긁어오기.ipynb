```python
# 오픈 API 아이디, 비번 (최초 설정 후 건드리지 말 것) 
client_id = "" 
client_secret = ""
```

```python
keyword_list = ['상품명 1', '상품명2', '상품명3']
```

```python
excel = openpyxl.Workbook()

worksheet1 = excel.active
worksheet1.title = '검색결과보기'
worksheet1.append(['keyword', 'title', 'lprice','hprice', 'productType', 'mallName', 'mallLink', 'imgLink', 'product_ID', 'brand', 'maker', 'category1', 'category2', 'category3', 'category4' ])
```

```python
i = 0

for keyword in keyword_list :    
    encText = urllib.parse.quote(keyword)
    url = "https://openapi.naver.com/v1/search/shop?query="+encText+"&display=100&start=1&sort=sim"
    # sim (유사도순), date (날짜순), asc(가격오름차순) ,dsc(가격내림차순)
    
    # url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode1==200):
        response_body = response.read().decode('utf-8')
    #     print(response_body1)
    else:
        print("Error Code:" + rescode)
    
    total = json.loads(response_body)['total']
    print('총 검색 수: ', total)

    total_page = int(total/100)+1

    if total_page >= 2 :
        for page in range(2, 10): 
            url = "https://openapi.naver.com/v1/search/shop?query="+encText+"&display=100&start={}&sort=sim".format(int(str(page-1)+'01'))
         # url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
            request = urllib.request.Request(url)
            request.add_header("X-Naver-Client-Id",client_id)
            request.add_header("X-Naver-Client-Secret",client_secret)
            response = urllib.request.urlopen(request)
            rescode = response.getcode()
            if(rescode1==200):
                response_body = response.read().decode('utf-8')
            #     print(response_body1)
            else:
                print("Error Code:" + rescode)
            
            item_json = json.loads(response_body)['items']


            for elm in item_json :

                #변수 저장
                title = elm.get('title').replace("<b>", "").replace("</b>","").replace("관부가세", "").replace("포함", "")
                lprice = int(elm.get('lprice'))
                hprice = int(elm.get('hprice'))
                mallName = elm.get('mallName')
                mallLink = elm.get('link')
                productType = int(elm.get('productType'))
                imgLink = elm.get('image')
                product_ID = elm.get('productId')
                brand = elm.get('brand')
                maker = elm.get('maker')
                category1 = elm.get('category1')
                category2 = elm.get('category2')
                category3 = elm.get('category3')
                category4 = elm.get('category4')

                worksheet1.append([keyword, title, lprice, hprice, mallName, mallLink, imgLink, product_ID, brand, maker, category1, category2, category3, category4 ])

                # 이미지 요청 및 다운로드
                urllib.request.urlretrieve(imgLink, 'C:/path/'+ product_ID + '.jpg')

    item_json = json.loads(response_body)['items']


    for elm in item_json :

        #변수 저장
        title = elm.get('title').replace("<b>", "").replace("</b>","").replace("관부가세", "").replace("포함", "")
        lprice = int(elm.get('lprice'))
        hprice = int(elm.get('hprice'))
        mallName = elm.get('mallName')
        mallLink = elm.get('link')
        productType = int(elm.get('productType'))
        imgLink = elm.get('image')
        product_ID = elm.get('productId')
        brand = elm.get('brand')
        maker = elm.get('maker')
        category1 = elm.get('category1')
        category2 = elm.get('category2')
        category3 = elm.get('category3')
        category4 = elm.get('category4')

        worksheet1.append([keyword, title, lprice, hprice, mallName, mallLink, imgLink, product_ID, brand, maker, category1, category2, category3, category4 ])

        # 이미지 요청 및 다운로드(경로설정할 것)
        urllib.request.urlretrieve(imgLink, 'C:/path'+ product_ID + '.jpg')

    i += 1
        
    print(i,':', keyword, 'crawling succeed.')
```
